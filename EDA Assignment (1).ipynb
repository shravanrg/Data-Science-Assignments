## Negative Value Detection & Fix

```python
neg_rows = df[(df < 0).any(axis=1)]
print("Total negative rows:", neg_rows.shape)
neg_rows.head()
```

```python
neg_counts=(df<0).sum()
print(neg_counts)
```

```python
cols_with_neg = neg_counts[neg_counts>0].index.tolist()
print("columns with negatives:", cols_with_neg)

# show first 10 rows that have any negative
df[(df<0).any(axis=1)].head(10)
```

```python
for col in cols_with_neg:
    # mark negative as NaN
    df.loc[df[col] < 0, col] = np.nan
    # fill these NaNs with median of that column
    med = df[col].median()
    df[col].fillna(med, inplace=True)
    print(f"Fixed column: {col}  (median used = {med})")
```

```python
# confirm no negatives remain and no nulls
print("\nNegative counts after fix:\n", (df < 0).sum())
print("\nMissing counts after fix (should be 0):\n", df.isnull().sum())
```

```python
# Quick sanity describe
print("\nDescribe after fixes:")
print(df.describe().T)
```

## Outlier Detection & Treatment

```python
# IQR outlier mask (per-row: any column is outlier)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outlier_mask = ((df < (Q1 - 1.5*IQR)) | (df > (Q3 + 1.5*IQR))).any(axis=1)
outlier_count = outlier_mask.sum()
print("Outlier rows (IQR rule):", outlier_count)

# Show a few outlier rows (for manual inspection)
if outlier_count>0:
    print("Sample outlier rows (first 5):")
    display(df[outlier_mask].head(5))

# Decision: drop if small, else cap at 1%/99%
if outlier_count <= 100:
    print("Dropping outlier rows (<=100).")
    df = df[~outlier_mask].reset_index(drop=True)
else:
    print("Capping values at 1% and 99% quantiles (winsorize).")
    lower = df.quantile(0.01)
    upper = df.quantile(0.99)
    df = df.clip(lower=lower, upper=upper, axis=1)
```

```python
print("Shape after outlier handling:", df.shape)
```
