```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("sales_data_with_discounts.csv")

# View first 5 rows
df.head()
```

```python
# Dataset information
df.info()

# Check missing values
df.isnull().sum()
```

```python
# Numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns
numerical_cols
```

```python
# Categorical columns
categorical_cols = df.select_dtypes(exclude=np.number).columns
categorical_cols
```

```python
# Mean
mean_values = df[numerical_cols].mean()

# Median
median_values = df[numerical_cols].median()

# Mode
mode_values = df[numerical_cols].mode().iloc[0]

# Standard Deviation
std_values = df[numerical_cols].std()

# Combine all statistics
stats_df = pd.DataFrame({
    "Mean": mean_values,
    "Median": median_values,
    "Mode": mode_values,
    "Standard Deviation": std_values
})

stats_df
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=20)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Box Plot of {col}")
    plt.show()
```

```python
for col in categorical_cols:
    plt.figure(figsize=(7,4))
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Bar Chart of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
#  Conclusion

Numerical variables show variation in sales and discount values

Presence of outliers suggests extreme sales/discount cases

Categorical analysis shows uneven category distribution

Dataset is now cleaned, visualized, and ready for advanced analytics or modeling
```
