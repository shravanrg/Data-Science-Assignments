### 1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# Create a list of 5 integers
numbers = [10, 20, 30, 40, 50]
print("Original list:", numbers)

# 1. Append - add an element to the end of the list
numbers.append(60)
print("After append(60):", numbers)

# 2. Extend - add multiple elements to the end
numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)

# 3. Insert - insert an element at a specific position
numbers.insert(2, 25)  # Insert 25 at index 2
print("After insert(2, 25):", numbers)

# 4. Remove - remove the first occurrence of a value
numbers.remove(40)
print("After remove(40):", numbers)

# 5. Pop - remove and return the last element (or specified index)
popped_item = numbers.pop()  # Removes last element
print("After pop():", numbers, " | Popped item:", popped_item)

# 6. Clear - remove all items from the list
numbers_copy = numbers.copy()  # Copy for further demonstration
numbers.clear()
print("After clear():", numbers)

# Restore the list for next operations
numbers = numbers_copy

# 7. Index - find the position of the first occurrence of a value
index_30 = numbers.index(30)
print("Index of 30:", index_30)

# 8. Count - count how many times a value appears
count_20 = numbers.count(20)
print("Count of 20:", count_20)

# 9. Sort - sort the list in ascending order
numbers.sort()
print("After sort():", numbers)

# 10. Reverse - reverse the elements of the list
numbers.reverse()
print("After reverse():", numbers)
```

### 7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python

# Input: take a number from the user
num = int(input("Enter a number: "))

# Check if the number is positive
if num > 0:
    print("The number is positive.")
    
    # Nested check: is it even?
    if num % 2 == 0:
        print("It is also even.")
    else:
        print("It is odd.")
        
elif num == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
```

### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
# Import Pandas
import pandas as pd
import numpy as np

# Step 1: Create a small DataFrame manually with 10 rows
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen', 'Ian', 'Jack'],
    'Age': [25, 30, 35, 40, 28, 33, 29, 31, 30, 25],
    'City': ['NY', 'LA', 'Chicago', 'NY', 'LA', 'Chicago', 'NY', 'LA', 'Chicago', 'NY'],
    'Salary': [50000, 60000, np.nan, 80000, 55000, np.nan, 70000, 65000, 60000, 50000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Step 2: Use .info() and .describe()
print("\nDataFrame Info:")
print(df.info())

print("\nDataFrame Description:")
print(df.describe())

# Step 3: Select 'Name' and 'City' columns
name_city = df[['Name', 'City']]
print("\nSelected Columns (Name and City):\n", name_city)

# Step 4: Drop 'City' column
df_dropped = df.drop(columns=['City'])
print("\nDataFrame after dropping 'City' column:\n", df_dropped)

# Step 5: Fill missing values in 'Salary' with mean
mean_salary = df['Salary'].mean()
df['Salary'].fillna(mean_salary, inplace=True)
print("\nDataFrame after filling missing Salary with mean:\n", df)

# Step 6: Remove any duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\nDataFrame after removing duplicates:\n", df_no_duplicates)
```
