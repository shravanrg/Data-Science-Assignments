# 2Ô∏è‚É£ Handling Missing Values
## üîπ Replace "?" with NaN

```python
df.replace("?", np.nan, inplace=True)
```

```python
df.isnull().sum()
```

## üîπ Handle missing values 

Categorical ‚Üí Mode

Numerical ‚Üí Median

```python
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

# 3Ô∏è‚É£ Scaling Techniques
## üîπ Separate numerical & categorical columns

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns

num_cols, cat_cols
```

## üîπ Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler_standard = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[num_cols] = scaler_standard.fit_transform(df[num_cols])
```

## üîπ Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler_minmax = MinMaxScaler()
df_minmax_scaled = df.copy()
df_minmax_scaled[num_cols] = scaler_minmax.fit_transform(df[num_cols])
```

# 4Ô∏è‚É£ Encoding Techniques
## üîπ Identify categorical columns with <5 categories

```python
low_cardinality = [col for col in cat_cols if df[col].nunique() < 5]
high_cardinality = [col for col in cat_cols if df[col].nunique() >= 5]

low_cardinality, high_cardinality
```

## üîπ One-Hot Encoding (Low Cardinality)

```python
df_encoded = pd.get_dummies(df, columns=low_cardinality, drop_first=True)
```

## üîπ Label Encoding (High Cardinality)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in high_cardinality:
    df_encoded[col] = le.fit_transform(df_encoded[col])
```

# 5Ô∏è‚É£ Feature Engineering
## üîπ Feature 1: Age Group

```python
df_encoded["age_group"] = pd.cut(
    df_encoded["age"],
    bins=[0,25,45,65,100],
    labels=["Young","Adult","Senior","Old"]
)

df_encoded["age_group"] = LabelEncoder().fit_transform(df_encoded["age_group"])
```

## üîπ Feature 2: Capital Gain Flag

```python
df_encoded.columns
```

```python
df_encoded["capital_gain_flag"] = (df_encoded["capital_gain"] > 0).astype(int)
```

```python
cap_col = [col for col in df_encoded.columns if "capital" in col.lower() and "gain" in col.lower()][0]

df_encoded["capital_gain_flag"] = (df_encoded[cap_col] > 0).astype(int)
```

# 6Ô∏è‚É£ Feature Transformation (Log Transformation)
## üîπ Check skewness

```python
df_encoded["capital_gain"].skew()
```

## üîπ Apply log transformation

```python
df_encoded["capital_gain_log"] = np.log1p(df_encoded["capital_gain"])
```
